```python
import xml.etree.ElementTree as ET
from xml.sax.saxutils import escape
# file='resources/horsies.xml'
file='resources/DMR20180718TCH.xml'
tree = ET.parse(file)
root = tree.getroot()
print(root)
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
for race in root.findall('RACE'):
    for breed in race.findall('BREED'):
        print(breed.text)
```

```python
import json
import xmltodict
 
with open(file, 'r') as f:
    xmlString = f.read()
 
# print("XML input (sample.xml):")
# print(xmlString)
     
jsonString = json.dumps(xmltodict.parse(xmlString), indent=4)
 
print("\nJSON output(output.json):")
print(jsonString)
```

```python
data=json.loads(jsonString)
data=data['CHART']['RACE']
distance=[]
start=[]
finish=[]
jockey=[]
trainer=[]
owner=[]
poc1l=[]
poc2l=[]
poc3l=[]
poc4l=[]
poc5l=[]
pocfl=[]
poc1p=[]
poc2p=[]
poc3p=[]
poc4p=[]
poc5p=[]
pocfp=[]
f1=[]
f2=[]
f3=[]
f4=[]
f5=[]
win=[]
for race in data:
#     print('Race: '+str(race['@NUMBER']))
    dist=race['DISTANCE']
    fr1=race['FRACTION_1']
    fr2=race['FRACTION_2']
    fr3=race['FRACTION_3']
    fr4=race['FRACTION_4']
    fr5=race['FRACTION_5']
    frw=race['WIN_TIME']
#     print('Distance: '+str(race['DISTANCE']))
    for entry in race['ENTRY']:
        distance.append(dist)
        f1.append(fr1)
        f2.append(fr2)
        f3.append(fr3)
        f4.append(fr4)
        f5.append(fr5)
        win.append(frw)
        start.append(entry['START_POSITION'])
        finish.append(entry['OFFICIAL_FIN'])
#         print('  Start Position: '+entry['START_POSITION'])
#         print('  Finish Position: '+entry['OFFICIAL_FIN'])
#         print('  Horse: '+entry['NAME'])
#         print('  Jockey: '+str(entry['JOCKEY']['FIRST_NAME'])+' '+str(entry['JOCKEY']['LAST_NAME']))
        jockey.append(str(entry['JOCKEY']['FIRST_NAME'])+' '+str(entry['JOCKEY']['LAST_NAME']))
#         print('  Trainer: '+str(entry['TRAINER']['FIRST_NAME'])+' '+str(entry['TRAINER']['LAST_NAME']))
        trainer.append(str(entry['TRAINER']['FIRST_NAME'])+' '+str(entry['TRAINER']['LAST_NAME']))
#         print('  Owner: '+str(entry['OWNER']))
        owner.append(entry['OWNER'])
#         print('  Points of Call:')
        for poc in entry['POINT_OF_CALL']:
            if(poc['@WHICH']=='1'):
                poc1l.append(poc['LENGTHS'])
                poc1p.append(poc['POSITION'])
            elif(poc['@WHICH']=='2'):
                poc2l.append(poc['LENGTHS'])
                poc2p.append(poc['POSITION'])
            elif(poc['@WHICH']=='3'):
                poc3l.append(poc['LENGTHS'])
                poc3p.append(poc['POSITION'])
            elif(poc['@WHICH']=='4'):
                poc4l.append(poc['LENGTHS'])
                poc4p.append(poc['POSITION'])
            elif(poc['@WHICH']=='5'):
                poc5l.append(poc['LENGTHS'])
                poc5p.append(poc['POSITION'])
            elif(poc['@WHICH']=='FINAL'):
                pocfl.append(poc['LENGTHS'])
                pocfp.append(poc['POSITION'])
        
```

```python
print(len(distance))
print(len(start))
print(len(finish))
print(len(jockey))
print(len(trainer))
print(len(owner))
print(len(poc1l))
print(len(poc2l))
```

```python
import pandas as pd

data={'distance':distance,
      'pp_start':start,
      'pp_finish':finish,
      'fraction1':f1,
      'fraction2':f2,
      'fraction3':f3,
      'fraction4':f4,
      'fraction5':f5,
      'fraction_win_time':win,
      'poc1':poc1l,
      'poc2':poc2l,
      'poc3':poc3l,
      'poc4':poc4l,
      'poc5':poc5l,
      'pocf':pocfl
     }
df = pd.DataFrame(data)
df.head(25)
```

```python
horseLength=8
furlong=660
dist1=1320
dist2=2640
dist3=3960
df['speed1']=((dist1-(horseLength*df['poc1'].astype(float)))/df['fraction1'].astype(float)).round(2)
df['speed2']=((dist2-(horseLength*df['poc2'].astype(float)))/df['fraction2'].astype(float)).round(2)
df['speed3']=((dist3-(horseLength*df['poc3'].astype(float)))/df['fraction3'].astype(float)).round(2)
df['speedf']=(((df['distance'].astype(float)/100*furlong)-(horseLength*df['pocf'].astype(float)))/df['fraction_win_time'].astype(float)).round(2)
df.head()
```
